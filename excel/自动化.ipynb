```python
from pathlib import Path

def read_excel_data(path):
    df = pd.read_excel(path, header=None)
    row = {}
    row['ID'] = df.iloc[3, 7]
    row['姓名'] = df.iloc[3, 1]
    row['开始日期'] = df.iloc[0, 7]
    row['结束日期'] = df.iloc[0, 9]
    total = df.loc[df[0] == '总计']
    row['酒店'] = total.iloc[0, 3]
    row['交通'] = total.iloc[0, 4]
    row['油费'] = total.iloc[0, 5]
    row['餐饮'] = total.iloc[0, 6]
    row['电话'] = total.iloc[0, 7]
    row['招待费'] = total.iloc[0, 8]
    row['杂项'] = total.iloc[0, 9]
    row['总计'] = total.iloc[0, 10]
    return row

def main():
    CWD = './Desktop/自动化练习表格'
    
    p = Path(CWD)
    rows = []
    for q in p.glob('**/[!All]*.xlsx'):
        rows.append(read_excel_data(q))
    df = pd.DataFrame(rows)
    writer = pd.ExcelWriter(p/'All.xls',engine='xlwt')
    df.to_excel(writer)
    writer.save()
    
if __name__ == '__main__':
    main()
```
